# Required Packages

```python
#Import required packages
import os
import librosa
import numpy as np
from tqdm import tqdm
import seaborn as sns
import matplotlib.pyplot as plt
from models import autoencoder_baseline
from tensorflow.keras.optimizers import Adam
from utils import generate_dataset, load_sound_file
from reconstruction import reconstruction
from detection import detection
```

# Load and Explore Dataset

```python
#Setting paths for normal and anomaly sound files
anomaly_path = "abnormal/"
anomaly_files = [os.path.join(anomaly_path, file) for file in os.listdir(anomaly_path)]
normal_path = "normal/"
normal_files = [os.path.join(normal_path, file) for file in os.listdir(normal_path)]

#Create test data files and labels
test_files = normal_files[-len(anomaly_files):] + anomaly_files
test_labels = np.hstack((np.zeros(len(anomaly_files)), np.ones(len(anomaly_files))))

#Training data files
train_files = normal_files[:-len(anomaly_files)]
```

```python
#Display normal and anomaly signals

#Load normal and anomaly sound signals
normal_signal, _ = load_sound_file(normal_files[0])
anomaly_signal, sr = load_sound_file(anomaly_files[0])

#Display the signals
plt.figure(figsize=(8,5),dpi=100)
sns.set_theme(style='whitegrid')
#Normal
librosa.display.waveshow(
    normal_signal,
    sr=sr,
    label="Normal Signal")
#Anomaly
librosa.display.waveshow(
    anomaly_signal,
    sr=sr,
    label="Anomaly Signal")
#Label the plots
plt.xlabel("Time (sec)")
plt.ylabel("Amplitude")
plt.legend();
```

# Feature Extraction and Training Dataset

```python
#Number of frequency bands
n_mels=64
#Time frame
frames=5
#Length of fft window
n_fft=1024
#Generate training dataset with feature extraction
train_data = generate_dataset(train_files, n_mels, frames, n_fft)
```

# Model Design and Compilation

```python
#Set model parameters
#Shape of the input data
input_shape = n_mels*frames
#Loss function
model_loss = 'mean_squared_error'

#Optimizer learning rate
lr = 1e-3

#Batch size and number of epochs to train the model
batch_size = 512

epochs = 30

#Create the baseline model and compile it with the hyperparameters
baseline_model = autoencoder_baseline(input_shape)

baseline_model.compile(loss=model_loss,
                       optimizer=Adam(learning_rate=lr))
#Print model summary
baseline_model.summary()
```

# Model Training

```python
#Model training
baseline_hist = baseline_model.fit(
    train_data, 
    train_data,
    batch_size=batch_size,
    epochs=epochs,
    verbose=2
)
```

```python
#Plot model loss
plt.figure(figsize=(8,5), dpi=100)
plt.plot(range(epochs), baseline_hist.history['loss'])
#Label the plot
plt.xlabel('Epochs')
plt.ylabel('mse loss')
plt.title('Model training loss')
```

# Signals Reconstruction

```python
#Perform reconstruction using the test data files and calculate mse error scores
reconstruction_errors = reconstruction(baseline_model,
                                       test_files,
                                       test_labels,
                                       n_mels,
                                       frames,
                                       n_fft)
```

# Anomaly Detection and Model Evaluation

```python
#Perform detection and evaluate model performance
detection(reconstruction_errors, test_labels)
```
