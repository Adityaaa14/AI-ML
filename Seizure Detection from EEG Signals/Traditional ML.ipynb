##Read dataset

```python
#Load data from csv
df = pd.read_csv('/content/gdrive/MyDrive/df_features.csv')
#Drop unnecessary columns
df=df.drop('Unnamed: 0',axis=1)
#Give column names
df.columns = feature_names
```

#Train/Test Split

```python
#Get all features
X = df.loc[:, df.columns != 'seizure']
#Get labels
y = df['seizure']
#Create train/test split with 80/20 ratio
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, shuffle = True, random_state=4)
```

#Metrics

```python
#Function to plot and label confusion matrix
def plot_confusion_matrix(validations, predictions):
  #Calculate confusion matrix
  matrix = confusion_matrix(validations, predictions)
  #Plot the matrix
  plt.figure(figsize=(5, 5), dpi=100)
  sns.set_theme(style='whitegrid')
  sns.heatmap(matrix,
              cmap='Blues',
              linecolor='white',
              linewidths=1,
              xticklabels=list(labels.keys()),
              yticklabels=list(labels.keys()),
              annot=True,
              fmt='d')
  #Label the matrix
  plt.title('Confusion Matrix')
  plt.ylabel('Ground Truth')
  plt.xlabel('Predictions')

#Function to plot feature importance
def feat_importance(coef, names, top):
  #Calculate feature importance score
  imp = abs(coef)
  imp,names = zip(*(sorted(zip(imp,names))[len(feature_names)-top:]))
  #Create and label the plot
  plt.figure(figsize=(10,5), dpi=100)
  sns.set_theme(style='whitegrid')
  plt.barh(range(len(names)), imp, align='center', height=0.4)
  plt.yticks(range(len(names)), names)
  plt.xlabel('Feature Coefficient')
  plt.show()
```

#Support Vector Machine

```python
#Create svm model
svm_model = SVC(kernel='linear', probability=True)
#Train the model
svm_model.fit(X_train, y_train)

#Make predictions
svm_pred = svm_model.predict(X_test)
svm_dec_sc = svm_model.decision_function(X_test)

#Calculate ROC and AUC
svm_fpr, svm_tpr, svm_thres = roc_curve(y_test, svm_dec_sc)
svm_auc = roc_auc_score(y_test, svm_dec_sc)
```

##Classification Report

```python
#Create classification report
print(classification_report(y_test, svm_pred, target_names=[l for l in list(labels.keys())]))
```

##Confusion matrix

```python
#Calculate and plot confusion matrix
plot_confusion_matrix(y_test, svm_pred)
```

##Feature importance

```python
#Plot feature importance
feat_importance(np.squeeze(svm_model.coef_), feature_names, 15)
```

##ROC Curve

```python
#Plot ROC curve
plt.figure(figsize=(10, 6), dpi=100)
plt.plot(svm_fpr, svm_tpr, label='SVM')
plt.plot([0,1],[0,1], 'k--', label='Random Guess')
#Label the plot
plt.xlabel("FPR")
plt.ylabel("TPR")
plt.legend(loc="best")
plt.title("ROC curve")

#AUC score
print("SVM AUC: ", svm_auc)
```

#XGBoost

```python
#Create XGB model
xgb_model = XGBClassifier()
#Train the model
xgb_model.fit(X_train, y_train)

#Make predictions
xgb_pred = xgb_model.predict(X_test)
xgb_dec_sc = xgb_model.predict_proba(X_test)[:,1]

#Calculate ROC and AUC scores
xgb_fpr, xgb_tpr, xgb_thres = roc_curve(y_test, xgb_dec_sc)
xgb_auc = roc_auc_score(y_test, xgb_dec_sc)
```

##Classification report

```python
#Create classification report
print(classification_report(y_test, xgb_pred, target_names=[l for l in list(labels.keys())]))
```

##Confusion matrix

```python
#Plot confusion matrix
plot_confusion_matrix(y_test, xgb_pred)
```

##Feature importance

```python
#Plot feature importance
fig, ax = plt.subplots(figsize=(10,5), dpi=100);
plot_importance(xgb_model,
                ax=ax,
                height=0.4,
                max_num_features=15,
                show_values=False,
                importance_type='weight');
```

##ROC curve

```python
#Plot ROC curve
plt.figure(figsize=(10, 6), dpi=100)
plt.plot(xgb_fpr, xgb_tpr, label='XG Boost')
plt.plot([0,1],[0,1], 'k--', label='Random Guess')
#Label the plot
plt.xlabel("FPR")
plt.ylabel("TPR")
plt.legend(loc="best")
plt.title("ROC curve")
#AUC score
print("XGB AUC: ", xgb_auc)
```
